# Numerical Data Ploting

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#setting the background of the plot
sns.set(style = 'darkgrid')
```

```python
#loading the data
tips = sns.load_dataset('tips')
tips
```

```python
#relplot stands for relational plot
#by default scatter plot
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker')
#if you want to separate the persons who have paid the bill are smoker or nonsmoker, we are passing the variable name to 'hue'
#hue differentiates the data with the help of colour (here it is blue & orange) 
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style = 'time')
#style differentiates the data with the help of shape(here it is dot & cross)
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, style = 'size')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'size')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style= 'time', size = 'size')
#size will give a bigger size or a smaller size of the shapes
```

```python
from numpy.random import randn
df = pd.DataFrame(dict(time = np.arange(500), value = randn(500).cumsum()))
df.head()
```

```python
#relational plot & line chart

sns.relplot(x = 'time', y = 'value', kind = 'line', data = df, sort = True)
#sort = True means sort the data and then try to print it 
```

```python
df2 = pd.DataFrame(randn(500, 2).cumsum(axis = 0), columns = ['time', 'value'])
df2.head()
```

```python
sns.relplot(x = 'time', y = 'value', kind = 'line', data = df2, sort = True)
```

```python
#without sorting
sns.relplot(x = 'time', y = 'value', kind = 'line', data = df2)
```

# Categorical Data Ploting

```python
tips.head()
```

```python
#finding highest bill paid on which day
sns.catplot(x = 'day', y = 'total_bill', data = tips) #on Sat a person has paid the highest amount
```

```python
#jitter shows dispersion. By default jitter = True
#if jitter = False, data will be shown in a line
sns.catplot(x = 'day', y = 'total_bill', data = tips, jitter = False)
```

```python
#how much tip is given by smoker & non smoker
sns.catplot(x = 'smoker', y = 'tip', data = tips, order = ['No', 'Yes'])
```

```python
#in which day we have received an exception amount of tip
sns.catplot(x = 'day', y = 'tip', data = tips, kind = 'box', hue = 'sex')   #on Sat
```

```python
#dodge = False merges the boxes. By default dodge = True
sns.catplot(x = 'day', y = 'total_bill', kind = 'box', data = tips, hue = 'sex', dodge = False)
```

```python
#diamond dataset
diamond = sns.load_dataset('diamonds')
diamond.head()
```

```python
sns.catplot(x = 'color', y = 'price', data = diamond, kind = 'box')
```

```python
#merging multiple plots in a single plot
c = sns.catplot(x = 'day', y = 'total_bill', kind = 'box', data = tips, hue = 'sex', dodge = False)
b = sns.swarmplot(x = 'day', y = 'total_bill', color = 'y', size = 3, data = tips, ax = c.ax)
```

```python
#importing titanic dataset
titanic = sns.load_dataset('titanic')

#count plot gives the count of a particular data
#palette helps to manage color intensity
sns.catplot(x = 'deck', kind = 'count', palette = 'ch:0.95', data = titanic, hue = 'class')
```

```python
sns.catplot(x = 'deck', kind = 'count', palette = 'ch:0.50', data = titanic, hue = 'class')
```

```python
sns.catplot(x = 'deck', kind = 'count', palette = 'ch:1.50', data = titanic, hue = 'class')
```

# Visualizing Distribution of the Data

```python
x = randn(100)
```

```python
#distribution plot
#if you want to know whether the dataset is normalized or not
#hist stands for histogram
sns.distplot(x, hist = True, bins = 30)
```

```python
sns.displot(x, bins = 30)
```

```python
sns.distplot(x, bins = 30)
```

```python
sns.histplot(x, bins = 30)
```

```python
x = randn(10000)   #randomly generated data are normalized in nature
sns.distplot(x, bins = 30)   #more data gives more perfect curve
```
