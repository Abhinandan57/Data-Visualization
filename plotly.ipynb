```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
from plotly.offline import iplot
import plotly as py
from plotly import tools as tls
```

```python
import cufflinks as cf
```

```python
print(py.__version__)
```

```python
py.offline.init_notebook_mode(connected = True)
```

```python
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.randn(100, 3), columns = ['A', 'B', 'C'])
df.head()
```

```python
df['A'] = df['A'].cumsum() + 20
df['B'] = df['B'].cumsum() + 20
df['C'] = df['C'].cumsum() + 20
df.head()
```

```python
plt.plot(df)
```

```python
df.iplot()
```

```python
df.iplot(x = 'A', y = 'B', mode = 'markers', size = 25)
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.iplot(kind = 'bar', x = 'sex', y = 'survived', title = 'Survived', xTitle = 'Sex', yTitle = 'Survived')
```

```python
titanic['sex'].value_counts()
```

```python
cf.getThemes()    #theme means background
```

```python
#changing the theme
cf.set_config_file(theme = 'henanigans')

#bar plot
df.iplot(kind = 'bar', barmode = 'stack', bargap = 0.5)
```

```python
#horizontal bar plot
df.iplot(kind = 'barh', barmode = 'stack', bargap = 0.5)
```

```python
#box plot : very important when outlier detection is concerned
df.iplot(kind = 'box')
```
