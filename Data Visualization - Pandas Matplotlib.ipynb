```python
import seaborn as sns

iris = sns.load_dataset('iris')
iris.head()
```

```python
#line plot (default plot)
ax = iris.plot(figsize = (15, 8), title = 'Iris Dataset')
ax.set_xlabel('Value')
ax.set_ylabel('Index')
```

```python
df = iris.drop(['species'], axis = 1)
df.head()
```

```python
df.iloc[0]
```

```python
#bar plot
df.iloc[0].plot(kind = 'bar')
```

```python
df.iloc[0].plot.bar()
```

```python
df.iloc[0].plot.barh()  #horizontal bar plot
```

```python
#histogram
iris['sepal_length'].plot.hist()
```

```python
iris.plot.hist(figsize = (20, 10))
```

```python
iris.plot(kind = 'hist', stacked = False, bins = 50, figsize = (20, 10))
```

```python
iris.plot(kind = 'hist', stacked = True, bins = 50, figsize = (20, 10))
```

```python
#diff() for calculating the nth discrete difference along the given axis. 
#If 'x' is the input array, then the first difference is given by out[i]=x[i+1]-a[i].
iris['sepal_width'].diff().plot(kind = 'hist', stacked = True, bins = 100, figsize = (20, 10))
```

```python
iris.plot(kind = 'hist', stacked = False, bins = 100, figsize = (20, 10), orientation = 'horizontal')
```
